# Preprocessing
## Missing Data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan,4],
                  'B':[5,np.nan,np.nan,8],
                  'C':[10,20,30,40]})

df
```

## Removing missing data

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

### Threshold (Require that many non-NA values.)

```python
df.dropna(thresh=2)
```

## Filling in missing data

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=0)
```

```python
df['A'].fillna(df['A'].mean())
```

```python
df.fillna(df.mean())
```

## Groupby

The groupby method : to group rows of data together and call aggregate functions

```python
df = pd.read_csv('./Datasets/Universities.csv')

df.head() # Show first N rows (N=5 by default)
```

Now .groupby() method can be used to group rows together based off of a **categorical** column. This column will then be reassigned to be the index.

Steps:

1. Choose a categorical column to group by
2. Choose the aggregation function. Recall an aggregation function should take multiple values and return a single value (e.g. max,min, mean, std, etc...)

```python
# returns a special groupby object waiting to have an aggregate method called on it!
df.groupby('Year')

print('Type: \n', type(df.groupby('Year').mean()))
df.groupby('Year').mean()
```

```python
df.groupby('Year').mean().sort_index(ascending=False)
```
