```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Visualizing Plots

```python
x = [0,1,2]
y = [100,200,300]

# add semicolon to hide matplotlib text output
plt.plot(x,y);
```

```python
# When running a .py file , plt.show() needs to be added at the end of the commands

plt.plot(x,y)
plt.show()
```

## Basic Tools

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})

housing
```

## Style Calls

One of the main reasons to learn the absolute basics is to see how the style interactions effect the API.

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y, color='purple')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red',marker='o',markersize=20,linestyle='--')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```
